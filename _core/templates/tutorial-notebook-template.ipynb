## 3. 핵심 개념

실습 전에 해결할 문제와 결과를 예상한다. 아래 예제는 입력 배열을 제곱한 뒤 scale을 적용하는 작은 함수로 code, graph와 결과 해석의 공통 구조를 보여준다. 실제 notebook에서는 이 설명을 해당 주제의 수식, data flow와 설계 이유로 교체한다.

## 4. API 확인

먼저 API signature, parameter와 return을 확인한다. 실제 tutorial에서는 `src/`의 public API를 import하고 type, shape, dtype과 상태 변화를 설명한다.

```python
def transform(values, scale=1.0):
    """Square values and multiply them by scale."""
    values = np.asarray(values, dtype=np.float32)
    return scale * values**2

print(transform.__doc__)
```

## 5. 단계별 실습

### 5.1. 입력과 예상 결과

입력은 -2부터 2까지의 연속값이다. 제곱 연산을 적용하면 모든 출력이 0 이상이고 원점을 기준으로 대칭일 것으로 예상할 수 있다.

```python
x = np.linspace(-2.0, 2.0, 101, dtype=np.float32)
y = transform(x, scale=1.0)

print(f"x.shape={x.shape}, x.dtype={x.dtype}")
print(f"y.shape={y.shape}, y.dtype={y.dtype}")
print(f"y.min={y.min():.4f}, y.max={y.max():.4f}")
```

### 5.2. 결과 검증

Shape와 dtype이 유지되고 출력이 음수가 아닌지 assertion으로 확인한다. Assertion은 graph만 보고 놓칠 수 있는 계약 위반을 자동으로 찾는다.

```python
assert y.shape == x.shape
assert y.dtype == np.float32
assert np.all(y >= 0.0)
assert np.isclose(transform(np.array([-2.0], dtype=np.float32))[0], 4.0)
print("validation passed")
```

## 6. 시각화

계산 결과를 graph로 확인한다. 축 이름, title, legend와 grid를 표시하여 graph만 보아도 무엇을 비교하는지 알 수 있게 한다.

```python
figure_path = OUTPUT_DIR / "transform-example.png"

fig, ax = plt.subplots(figsize=(7, 4))
ax.plot(x, y, label="scale=1.0", linewidth=2)
ax.set_title("Transform Output")
ax.set_xlabel("input")
ax.set_ylabel("output")
ax.grid(alpha=0.3)
ax.legend()
fig.tight_layout()
fig.savefig(figure_path, dpi=120)
plt.show()

assert figure_path.exists()
print(f"saved={figure_path}")
```

## 7. 결과 해석과 설계 이유

Graph는 예상대로 원점을 기준으로 대칭이고 출력이 0 이상임을 보여준다. `np.asarray(..., dtype=np.float32)`를 사용한 이유는 입력 type과 관계없이 계산 dtype을 고정하여 결과를 재현하기 위해서다. 실제 tutorial에서도 결과를 나열하는 데 그치지 않고 왜 그런 결과가 나왔는지와 구현 선택의 이유를 설명한다.

## 8. 추가 과제

Scale을 0.5로 바꾸어 출력 크기의 변화를 확인한다. 추가 과제도 기본 상태에서 실행 가능하게 작성하고, 학습자가 수정할 지점은 Markdown으로 안내한다.

```python
y_half = transform(x, scale=0.5)
assert np.allclose(y_half, 0.5 * y)

fig, ax = plt.subplots(figsize=(7, 4))
ax.plot(x, y, label="scale=1.0")
ax.plot(x, y_half, label="scale=0.5")
ax.set_title("Scale Comparison")
ax.set_xlabel("input")
ax.set_ylabel("output")
ax.grid(alpha=0.3)
ax.legend()
fig.tight_layout()
plt.show()
```
